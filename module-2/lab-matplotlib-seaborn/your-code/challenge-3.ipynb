#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
fig, ax = plt.subplots(3,1, figsize=(12,6))
titanic.plot(y='Age', kind='hist',ax = ax[0], bins=10)
titanic.plot(y='Age', kind='hist',ax = ax[1], bins=20)
titanic.plot(y='Age', kind='hist',ax = ax[2], bins=50)
```

#### Whether the bin size affects your plot? Why?

They did affect how the plots look because the less bins you have the less you will segregated your data is. This could be possitibe for getting an general overview of the data however you lose in accuracy of the data. In this specific case, the bins=20 seems to be most appropiate between the three, since it gives enough information about it individual bin while working properly with the adjacent bins.

#### What does the previous plot tell you about the feature `Age`? Comment.

We can see that the median of Age is around the value 29 being the 50% per cent of the values between the lower quartile at around 22 and the upper quartile at around 37, also known as the interquartile range. The lower fence is at 0, and the upper fence (1.5*IQR) is at around 59. Being the values above this fence considered outliers.

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
colormap = plt.cm.plasma_r
corr=titanic.corr()
sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
```

#### According to the previous plot, what are the most correlated feature? Comment.

The most correlated feature is Pclass and Fare which hold a negative correlation of 0.56, this is because in terms of classes, the lower the value the higher the class and therefore, the higher the fare.
