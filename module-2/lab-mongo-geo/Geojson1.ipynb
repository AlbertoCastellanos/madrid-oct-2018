## Phase 1: Getting coordinates and transform to geojason.

Importing the necessary packes

```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize
```

Connecting to mongo

```python
client=MongoClient("mongodb://admin:admin@localhost:32772")
db=client.ccc
```

Querying the dataset to find all companies that have complete information about location and number of employees

```python
query1=db.companies.find({"offices.latitude":{"$ne":None, "$exists":True},"offices.longitude":{"$ne":None, "$exists":True}, "number_of_employees":{"$ne":None, "$exists":True} },projection={"name":True, "number_of_employees":True, "_id":False, "offices": True})
df1= pd.DataFrame(list(query1))
print(len(df1))
#df1
#print(df1['offices'])
#df1['offices']= [e for e in df1['offices']]
#[e for e in df1['offices']]
#df1['offices'][0][0]["latitude"]
```

Adding independent columns for latitude and longitude and removing the object column

```python
df1['latitude']=[e[0]["latitude"] for e in df1['offices']]
df1['longitude']=[e[0]["longitude"] for e in df1['offices']]

df1=df1[['name', 'number_of_employees', 'latitude', 'longitude']]
df1
```

Transforming coordinates to geojason format

```python
locations=[]

for e in range(len((df1['latitude']))):
    
    locations.append({
    'type': 'Point',
    'coordinates': [
        df1.iloc[e]['longitude'],
        df1.iloc[e]['latitude']
    ]
    }
    )
```

Cleaning out the dataframe

```python
df1Clean=df1[['name', 'number_of_employees']]
df1Clean['location']=locations
df1Clean
```

Exporting the dataset to json

```python
with open('filename.json', 'w')  as f:
   for i, e in df1.iterrows():
       f.write(e.to_json()+'\n')
```

After exporting to json, the new data set is loaded in Compass as a new collection. Once loaded, the index of the location field needs to be set to 2dSphere in order to be able to query the geojason.

## Phase 2: Link to the new dataset and query by radius

A new cnnection is defined to connect with the new dataset

```python
client=MongoClient("mongodb://admin:admin@localhost:32772")
db=client.ccc
```

A function is defined to make geo queries easier to write

```python
def geoQueryNear(geoJsonField, lat, lng, distance):
    return {
        geoJsonField: {
            "$near": {
                "$geometry": {
                    "type": "Point", 
                    "coordinates": [lng, lat]
                },
                "$maxDistance":distance
            }
        }
    }

#result = db.GeoComp.find(geoQueryNear('location',40.4799741, -3.7360907, 100000))
#list(result)
```

Another function is defined to apply the queries and to loop through the original dataframe's coordinates

```python
def getEmployeesInRadius(df1,Radius):
    result=[]
    for e in range(len((df1['latitude']))):
        result.append(list(db.GeoComp.find(geoQueryNear('location', df1.iloc[e]['latitude'],df1.iloc[e]['longitude'], Radius))))
    Totnum=0
    TotalNumberOfEmployees=[]
    for e in result:
        for x in e: 
            Totnum+=x['number_of_employees']
        TotalNumberOfEmployees.append(Totnum)
        Totnum=0
    df2=df1
    df2['TotalEmployees']=TotalNumberOfEmployees
    return df2.sort_values(['TotalEmployees'], ascending=0).head(10)
```

Getting companies that have a largest number of employees in a radius of 100 meters.

```python
a=getEmployeesInRadius(df1,100)
a
```

Getting companies that have a largest number of employees in a radius of 1000 meters.

```python
b=getEmployeesInRadius(df1,1000)
b
```

Getting companies that have a largest number of employees in a radius of 10000 meters.

```python
c=getEmployeesInRadius(df1,10000)
c
```

Exporting all results to json in order to show them graphically in Tableau

```python
a.to_json('./LargestEmployeeRatio100.json', orient='records')
b.to_json('./LargestEmployeeRatio1000.json', orient='records')
c.to_json('./LargestEmployeeRatio10000.json', orient='records')
```
