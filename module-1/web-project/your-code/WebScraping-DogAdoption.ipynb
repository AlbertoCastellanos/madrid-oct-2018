# Web Project - Web Scrapping

## Dog Rescue - Scrapping information about available dogs for adoption in Madrid

The current project aims to retrieve information about dogs in adoption from different websites owned by animal shelters in Madrid in order to have a unique list that contains all dogs that are available for adoption in Madrid.

```python
import requests
from bs4 import BeautifulSoup
from lxml import html
from lxml.html import fromstring
import random
import pandas as pd
```

```python
sitio1url1="https://nuevavida-adopciones.org/tipo-animal/perro/"
```

```python
reqSitio1=requests.get(sitio1url1).content
soupSitio1 = BeautifulSoup(reqSitio1, "lxml")
print(soupSitio1.prettify())
```

```python
numDogs=[e for e in soupSitio1.select('.name  a')]
numDogs
```

```python
names=[e.text for e in numDogs]
links=[e['href'] for e in numDogs]
```

```python
numDogs=[e for e in soupSitio1.select('.name  a')]
```

```python
picDogs=[e['src'] for e in soupSitio1.select('.attachment-woocommerce_thumbnail')]
```

```python
Web1=pd.DataFrame({"images": picDogs, "names": names,"links":links})
Web1.head()
```

# SITIO 2

```python
sitio2url1="http://arcadenoe.org/listado_animales.php?id_seccion=25&p=1"
```

```python
reqSitio2=requests.get(sitio2url1).content
soupSitio2 = BeautifulSoup(reqSitio2, "lxml")
print(soupSitio2)
```

```python
Dogs2=[e for e in soupSitio2.select('.c_l_contenedor_imagen a')]
#names2=[e.text for e in Dogs2]
links2=[e['href'] for e in Dogs2]
```

```python
links2
```

```python
names2=[e.text for e in soupSitio2.select('.cuadro_listado h2')]
names2
```

```python
picDogs2=[e['src'] for e in soupSitio2.select('.c_l_contenedor_imagen a img')]
domain="http://arcadenoe.org"
picDogs2= [(domain+e) for e in picDogs2] 
picDogs2
```

```python
Web2=pd.DataFrame({"images": picDogs2, "names": names2,"links":links2})
Web2.head()
```

# RESULT

```python
result=pd.concat([Web1, Web2])
```
