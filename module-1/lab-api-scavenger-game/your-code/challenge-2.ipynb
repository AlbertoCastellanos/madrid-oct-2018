```python
import json
import requests
import pandas as pd
```

All necessary modules have been imported, now we will requeste the commit from 20th Oct to 25th Oct and will expand the page limit to 100 registers. After getting the data, we convert it to json

```python
reqCommits=requests.get('https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/commits?since=2018-10-20T00:00:00Z&until=2018-10-25T23:59:59Z&per_page=100')
commits=reqCommits.json()
commits
```

Once the json is produced, we allocate the data in a dataframe and check its columns

```python
dataCommits=commits
dfCommits=pd.DataFrame(dataCommits)
dfCommits.columns
```

We know from studying the API documentation that the information we aim for is within a subarray in "commit". We generate the flatten function to normalize the subarray and then pass the new data into a new dataframe called flat

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data

flat=flatten(dfCommits, ["commit"])
flat.columns
```

We need to flatten the dataframe once more:

```python
flat=flatten(flat,['commit_author','commit_committer'])
flat.columns
```

We will define a comprehensive list to obtain a list of values matching the given time period

```python
result=([i for i in flat['commit_author_date']])
```
